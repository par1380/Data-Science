# Data review

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("E:\Parsa\project of data analytics\project2(Demand analysis and forecasting with classic machine learning models)\simulated_sales_data.csv")
data.head()
```

```python
data.describe()
```

```python
df = pd.DataFrame(data)

variance_data = {"var_price":df["Price"].var() , "var_discount": df["Discount"].var() , "var_advertising":df["Advertising"].var() , "var_sales":df["Sales"].var()}
variance_data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler_df = scaler.fit_transform(df[["Price" , "Discount" , "Advertising" , "Sales"]])
scaler_df = pd.DataFrame(scaler_df , columns=["Price" , "Discount" , "Advertising" , "Sales"])
scaler_df = pd.concat([ df[["Category","Season"]] , scaler_df] , axis = 1)

scaler_df
```

## Linear_model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error

unique_category = df["Category"].unique()

category_models = {}

for cat in unique_category:
    
    filtered_df = scaler_df[scaler_df["Category"]==cat]
    
    X = filtered_df[["Price" , "Discount" , "Advertising" ]]
    Y = filtered_df[["Sales"]]
    x_train , x_test , y_train , y_test = train_test_split(X,Y , test_size=0.2 , random_state=42)
    linearregression = LinearRegression()
    linearregression.fit(x_train,y_train)
    
    category_models[cat] = linearregression
    
    y_hat = linearregression.predict(x_test)
    mse_linear = mean_squared_error(y_test , y_hat)
    r2_linear = r2_score(y_test , y_hat)
    mae_linear = mean_absolute_error(y_test , y_hat)
    
    print(cat)
    print(f" this is mean_absolut_error :{mae_linear}")
    print(f" this is mean_squared_error :{mse_linear}")
    print(f" this is r2_score :{r2_linear}\n")
```

## Desicion Tree Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import  mean_squared_error , r2_score , mean_absolute_error


unique_category = df["Category"].unique()

category_models = {}

for cat in unique_category:
    
    filtered_df = scaler_df[scaler_df["Category"]==cat]
    
    X = filtered_df[["Price" , "Discount" , "Advertising" ]]
    Y = filtered_df[["Sales"]]
    x_train , x_test , y_train , y_test = train_test_split(X,Y , test_size=0.2 , random_state=42)
    model = DecisionTreeRegressor()
    model.fit(x_train,y_train)
    
    category_models[cat] = model
    
    y_hat = model.predict(x_test)

    mse_tree = mean_squared_error(y_test , y_hat)
    r2_tree = r2_score(y_test , y_hat)
    mae_tree = mean_absolute_error(y_test , y_hat)
    
    print(cat)
    print(f" this is mean_absolut_error :{mae_tree}")
    print(f" this is mean_squared_error :{mse_tree}")
    print(f" this is r2_score :{r2_tree}\n")
```

## KNN Model

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import  mean_squared_error , r2_score , mean_absolute_error


unique_category = df["Category"].unique()

category_models = {}

for cat in unique_category:
    
    filtered_df = scaler_df[scaler_df["Category"]==cat]
    
    X = filtered_df[["Price" , "Discount" , "Advertising" ]]
    Y = filtered_df[["Sales"]]
    x_train , x_test , y_train , y_test = train_test_split(X,Y , test_size=0.2 , random_state=42)
    model = KNeighborsRegressor()
    model.fit(x_train,y_train)
    
    category_models[cat] = model
    
    y_hat = model.predict(x_test)

    mse_knn = mean_squared_error(y_test , y_hat)
    r2_knn = r2_score(y_test , y_hat)
    mae_knn = mean_absolute_error(y_test , y_hat)
    
    print(cat)
    print(f" this is mean_absolut_error :{mae_knn}")
    print(f" this is mean_squared_error :{mse_knn}")
    print(f" this is r2_score :{r2_knn}\n")
    
```

## Feature Engineering

```python
scaler_df["final_price"] = scaler_df["Price"] * (1-scaler_df["Discount"])

scaler_df["ad_budjet_per_sales"] = scaler_df["Advertising"]/scaler_df["Sales"]
scaler_df
```

## Optimazed LinearRgression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error

unique_category = df["Category"].unique()

category_models = {}

for cat in unique_category:
    
    filtered_df = scaler_df[scaler_df["Category"]==cat]
    
    X = filtered_df[["Price" , "Discount" , "Advertising","final_price" , "ad_budjet_per_sales"]]
    Y = filtered_df[["Sales"]]
    x_train , x_test , y_train , y_test = train_test_split(X,Y , test_size=0.2 , random_state=42)
    linearregression = LinearRegression()
    linearregression.fit(x_train,y_train)
    
    category_models[cat] = linearregression
    
    y_hat = linearregression.predict(x_test)
    mse_linear = mean_squared_error(y_test , y_hat)
    r2_linear = r2_score(y_test , y_hat)
    mae_linear = mean_absolute_error(y_test , y_hat)
    
    print(cat)
    print(f" this is mean_absolut_error :{mae_linear}")
    print(f" this is mean_squared_error :{mse_linear}")
    print(f" this is r2_score :{r2_linear}\n")
```
